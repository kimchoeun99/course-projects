```python
def residual_plot(y_pred, y):
    plt.scatter(y_pred, y_pred-y, c="blue", marker='s', label="Residual plot")
    plt.xlabel("Predicted values")
    plt.ylabel("Residuals")
    plt.hlines(y=0, color="black", xmin=-10, xmax=1000)
```

```python
def price_residuals(y_pred, y):
    plt.scatter(y, y_pred-y, c='blue', marker='s', label='price and residuals')
    plt.xlabel('actual price')
    plt.ylabel('residuals')
    plt.hlines(y=0, color='black', xmin=-10, xmax=1000)
```
